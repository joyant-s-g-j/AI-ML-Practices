```python
arr = [1, 2, 3]
print(arr * 5)
arr = [x*5 for x in arr]
print(arr)
```

```python
import numpy as np
ar = [1, 2, 3]
ar1 = np.array(ar)
ar1 = ar1*5
print(ar1)
print(type(ar1))

#2d
arr = [[1, 2, 3],
       [4, 5, 6]]
ar2 = np.array(arr)

#3d

ar3 = np.array([
    [[1, 2, 3],
     [4, 5, 6]],
    [[1, 2, 3],
     [4, 5, 6]],
    [[1, 2, 3],
     [4, 5, 6]]
    ])
```

```python
#dimension
print(ar1.ndim)
print(ar2.ndim)
print(ar3.ndim)

#shape
print(ar1.shape)
print(ar2.shape)
print(ar3.shape)

#data type
print(ar1.dtype)
print(ar2.dtype)
print(ar3.dtype)

#size
print(ar1.size)
print(ar2.size)
print(ar3.size)
```
