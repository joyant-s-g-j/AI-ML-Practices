```python
import numpy as np
import pandas as pd
```

```python
data = [45, 48, 50, 52, 55, 58, 60, 62, 65, 68, 70, 72, 75, 78, 80, 85, 90, 95, 150, 500]

df = np.array(data)
```

```python
data_range = np.max(df) - np.min(df)

print("Data Range: ", data_range)
```

```python
pop_variance = np.var(df, ddof=0)
smp_variance = np.var(df, ddof=1)

print("Population Variance: ", pop_variance)
print("Sample Variance: ", smp_variance)
```

```python
pop_std = np.std(df, ddof=0)
smp_std = np.std(df, ddof=1)

print("Population Standard deviation: ", pop_std)
print("Sample Standard deviation :", smp_std)
```

```python
q2 = np.median(df)

print("Q2 Percentile: ", q2)
```

```python
q1 = np.percentile(df, 25)
q3 = np.percentile(df, 75)

iqr = q3-q1

print("Standard deviation :", iqr)
```

```python
lf = q1 - 1.5 * iqr
uf = q3 + 1.5 * iqr

print("Lower Fence :", lf)
print("Upper Fence :", uf)
```

```python
outliers = df[(df < lf) | (df > uf)]

print("Outliers Count:", len(outliers))
print("Outliers :", outliers)
```
