```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({'h':[150, 160, 170, 180, 190],
                    'w': [50,60,70,80, 90]})

df
```

```python
m = df.mean()
s = df.std()

z = (df-m)/s
z.round(2)
```

```python
z.plot(kind='bar')
plt.title('Z scaling features')
plt.xlabel("Index")
plt.ylabel("Z value")
```
