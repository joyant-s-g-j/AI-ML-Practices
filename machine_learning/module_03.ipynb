```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.DataFrame({'h':[150, 160, 170, 180, 190],
                    'w': [50,60,70,80, 90]})

df
```

```python
m = df.mean()
s = df.std()

z = (df-m)/s
z.round(2)
```

```python
z.plot(kind='bar')
plt.title('Z scaling features')
plt.xlabel("Index")
plt.ylabel("Z value")
```

```python
mn = df.min()
mx = df.max()
rg = mx - mn

ss = df - mn

mm = ss/rg

mm.round(2)
```

```python
mm.plot(kind='bar')
plt.title('Min-Max scaling')
plt.xlabel("Index")
plt.ylabel("MIn-Max value")
```

```python
md = df.median()
q1 = df.quantile(.25)
q3 = df.quantile(.75)
iqr = q3 - q1

ct = df - md

rb = ct/iqr
rb.round(2)
```

```python
rb.plot(kind='bar')
plt.title('Robust scaling')
plt.xlabel("Index")
plt.ylabel("Robust value")
```
