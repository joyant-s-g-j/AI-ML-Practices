```python
a = np.random.randint(1, 10, size=(2, 3))
b = np.random.randint(20, 30, size=(2, 3))

print(a)
print(b)
```

```python
con_row = np.concatenate((a, b), axis=0)

print(con_row)

con_col = np.concatenate((a,b), axis=1)
print(con_col)
```

```python
x = np.array([10, 20, 30, 100])
y = np.array([2, 3, 4, 5])

add = x+2
print(add)

sin_val = np.sin(x)
print(sin_val)

deg_con = np.rad2deg(x)
print(deg_con)

base_10 = np.log10(x)
print(base_10)

sqrt = np.sqrt(x)
print(sqrt)

s = np.sum(x)
print(s)

cusum = np.cumsum(x)
print(cusum)

sq = np.square(y)
print(sq)
```

```python
x = np.array([[30, 10, 20],
              [50, 60, 40]])

hs = np.sort(x, axis=1)
print(hs)

vs = np.sort(x, axis=0)
print(vs)
```
