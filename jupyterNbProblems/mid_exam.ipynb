```python
import numpy as np
import pandas as pd

np.random.seed(42)

ids = np.arange(1, 11)
ages = np.random.randint(18, 60, 10)
salaries = np.random.randint(30000, 90000, 10)
departments = np.array(["HR", "IT", "Finance", "IT", "HR", "Sales", "Finance", "IT", "Sales", "HR"])

DF = pd.DataFrame({
  "id": ids,
  "age": ages,
  "salary": salaries,
  "dept": departments
})

DF.to_csv("employees.csv", index=False)
print("Sample Data Created and Saved as employees.csv")
```

### **NumPy Problem Solving**

## **Ans to the question number 1**

```python
ages = []
slry = []

with open('employees.csv', 'r') as f:
  lines = f.readlines()

head = lines[0].strip().split(',')
age_idx = head.index('age')
slry_idx = head.index('salary')

for line in lines[1:]:
  prt = line.strip().split(',')
  ages.append(int(prt[age_idx]))
  slry.append(int(prt[slry_idx]))

ages_arr = np.array(ages)
slry_arr = np.array(slry)

print(f"Ages array: { ages_arr }")
print(f"Ages array Data type: {ages_arr.dtype}, Dimension: {ages_arr.ndim}, Shape: {ages_arr.shape}, Size: {ages_arr.size}\n")

print(f"Salary array: { slry_arr }")
print(f"Salary array Data type: {slry_arr.dtype}, Dimension: {slry_arr.ndim}, Shape: {slry_arr.shape}, Size: {slry_arr.size}")
```

## **Ans to the question number 2**

```python
ages_arr = np.array(ages)
slry_arr = np.array(slry)

high = np.max(slry_arr)
low = np.min(slry_arr)

age_avg = np.mean(ages_arr)
slry_avg = np.mean(slry_arr)

print(f"Highest Salary: {high}")
print(f"Lowest Salary: {low}")
print(f"Average Age: {age_avg:.2f}")
print(f"Average Salary: {slry_avg:.2f}")
```

## **Ans to the question number 3**

```python
fltr_age = ages_arr[ages_arr > 30]
emp_count = fltr_age.size
print(f"Ages of all employees greater than 30: {fltr_age}\n")
print(f"Number of employees are older than 30: {emp_count}")
```

## **Ans to the question number 4**

```python
new_age = ages_arr + 5

print(f"Original ages of all employees: {ages_arr}")
print(f"Ages of all employees after adding 5 years: {new_age}")
```

## **Ans to the question number 5**

```python
total = np.sum(slry_arr)

dif = high - low

print(f"Total salary expense: {total}")
print(f"Difference between the maximum and minimum salary: {dif}")
```
