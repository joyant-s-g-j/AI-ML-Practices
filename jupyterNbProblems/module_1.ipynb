1-3 - Fundamental Data Structure

```python
import math

class Shape():
  def area(self):
    return 0

class Rectangle(Shape):
  def __init__(self, w, h):
    self.w = w
    self.h = h

  def area(self):
    return f"Area: {self.w * self.h:.2f}"

class Circle(Shape):
  def __init__(self, r):
    self.r = r

  def area(self):
    return f"Area: {math.pi * math.pow(self.r, 2):.2f}"

shp_info = input().split()
shp_type = shp_info[0]

if shp_type == "Rectangle":
  w, h = int(shp_info[1]), int(shp_info[2])
  shp = Rectangle(w, h)
  print(shp.area())

elif shp_type == "Circle":
  r = int(shp_info[1])
  shp = Circle(r)
  print(shp.area())
```

4 - Slice and Loop

```python
nums = [10, 20, 30, 40, 50, 60]
print(nums[2:4])    # middle slice
print(nums[-3:])    # last 3
print(nums[::2])    # step slice
print(nums[::-1])   # reverse
print(nums[-3:-1])  # here -1 means the previous index of last index
```

5 - Nested Data Structure

```python
data = {
    "user": {
        "name": "joyant",
        "id": 123,
        "contact" : {
            "email": ["joy@gmail.com", "joyant@gmail.com"],
            "phone" : "124-345"
        }
    },
    "items": [
        {"name": "Laptop", "price": 1200},
        {"name": "keyboard", "price": 75}
    ]
}

print("User name:", data["user"]["name"])
print("Item price:", data["items"][0]["price"])
print("User email:", data["user"]["contact"]["email"][1])
```

```python
order_data = {
    "customer": {
        "first_name": "Joyant",
        "last_name": "Sheikhar"
    },
    "items": [
        { "item_name": "Book", "price": 200 },
        { "item_name": "Pen", "price": 100 },
        { "item_name": "Bag", "price": 1200 },
    ]
}

full_name = f"{order_data['customer']['first_name']} {order_data['customer']['last_name']}"
print('Customer name:', full_name)

total_price = sum(item['price'] for item in order_data["items"])
print("Total Order Price:", total_price)

item_names = [item["item_name"] for item in order_data['items']]
print("Item names:", item_names)
```
