```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import csv

data = []

with open('Practice_Day_15_5.csv', 'r', newline='', encoding='utf-8') as f:
    reader = csv.reader(f)
    header = next(reader)
    n_cols = len(header)
    for row in reader:
        while len(row) < n_cols:
            row.append(pd.NA)
        if len(row) > n_cols:
            row = row[:n_cols]
        data.append(row)

df = pd.DataFrame(data, columns=header)
df
```

```python
df['Price'] = pd.to_numeric(df['Price'])

dis = df['Price'] * 0.10

df['Discounted_Price'] = df['Price'] - dis

df
```

```python
df.to_csv('discounted_products.csv')
```
